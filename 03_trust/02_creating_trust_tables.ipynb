## 1 Â· Understand Trust List Anatomy

A trust list is a YAML document with three mandatory keys:

| Key | Description |
|-----|-------------|
| `owner_name` | A human-readable name for the list owner |
| `attesters`  | Addresses/entities you trust to label addresses |
| `attestations` | Specific label claims you endorse |

Each **attester** carries a `weight` (0.0â€“1.0) that controls how much their
labels contribute to confidence scores transitively.

> You have to trust yourself; otherwise the trust chain has nowhere to start.

## 3 Â· Check Trust List for OLI Compliance

```python
import yaml

# Load your trust list
with open("trust_list.yaml", "r") as f:
    trust_list = yaml.safe_load(f)

owner_name   = trust_list.get("owner_name")
attesters    = trust_list.get("attesters",    [])
attestations = trust_list.get("attestations", [])

# Check if the trust list is OLI compliant
is_valid = oli.validate_trust_list(owner_name, attesters, attestations)
print("\nYour trust list is OLI compliant:", is_valid)
```

## 3.1 Â· (Alternatively) Want to Keep Your Trust List Private?

Not ready to publish? You can add your trust list locally using `oli.add_trust_list()`.
This loads it into your session only; it will not be visible to others or affect
the global trust graph. Useful for testing, back-testing, or private environments
where you want the benefits of trust scoring without sharing your endorsements.

```python
# add trust list to your local OLI instance
your_eoa = "0x0000000000000000000000000000000000000000"
oli.add_trust_list(owner_name, attesters, attestations, your_eoa)
```
